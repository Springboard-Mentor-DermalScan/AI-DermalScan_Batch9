```python
import random
import os
import cv2
import matplotlib.pyplot as plt

def show_random_images(class_name, num=4):
    folder = f"dataset/{class_name}"
    
    # Check if folder exists
    if not os.path.exists(folder):
        print(f"Error: Folder '{folder}' does not exist.")
        return
    
    # Get list of images
    images = os.listdir(folder)
    
    # Check if there are any images in the folder
    if len(images) == 0:
        print(f"Error: No images found in '{folder}'.")
        return
    
    # Adjust num if there are fewer images than requested
    if len(images) < num:
        print(f"Warning: Only {len(images)} images available. Showing all of them.")
        num = len(images)
    
    plt.figure(figsize=(10,5))
    for i in range(num):
        img_path = os.path.join(folder, random.choice(images))
        img = cv2.imread(img_path)
        
        # Check if image was loaded successfully
        if img is None:
            print(f"Warning: Could not load image {img_path}")
            continue
            
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.subplot(1, num, i+1)
        plt.imshow(img)
        plt.title(class_name)
        plt.axis("off")
    
    plt.tight_layout()
    plt.show()

# Make sure the folder exists and contains images before calling this function
show_random_images("wrinkles")
```

```python
import os
import pandas as pd

# Define the dataset path
dataset_path = "dataset"

# Get only the directories (not files) in the dataset folder
all_dirs = [d for d in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, d))]

# Create a dictionary to store the count of files in each class folder
data_count = {}

for c in all_dirs:
    class_path = os.path.join(dataset_path, c)
    # Count only files (not directories) in each class folder
    files = [f for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f)) and not f.startswith('.')]
    count = len(files)
    
    # Only add non-empty folders to our dictionary
    # Or if we already have this class, take the maximum count
    if count > 0:
        if c in data_count:
            data_count[c] = max(data_count[c], count)
        else:
            data_count[c] = count

# Display the counts
print("Number of images in each class:")
for class_name, count in data_count.items():
    print(f"{class_name}: {count} images")

# Convert to pandas DataFrame for a nicer display in Jupyter
df_counts = pd.DataFrame(list(data_count.items()), columns=['Class', 'Image Count'])
df_counts  # This will display as a table in Jupyter
```

```python
import matplotlib.pyplot as plt

# Make sure data_count only has unique keys
# Convert to a list of tuples for plotting
classes = list(data_count.keys())
counts = list(data_count.values())

# Create the bar chart with clear figure
plt.figure(figsize=(10, 6))  # Set figure size for better visibility
plt.clf()  # Clear the current figure to avoid overlapping plots

# Create the bar chart
plt.bar(classes, counts, color='skyblue')
plt.title("Class Distribution", fontsize=14)
plt.xlabel("Classes", fontsize=12)
plt.ylabel("Image Count", fontsize=12)

# Rotate x-axis labels if needed for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout to make sure everything fits
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
```

```python
IMG_SIZE = 224
classes = ["wrinkles", "dark_spots", "puffy_eyes", "clear_skin"]

label_dict = {
    "wrinkles": 0,
    "dark_spots": 1,
    "puffy_eyes": 2,
    "clear_skin": 3
}
```

```python
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img / 255.0  # normalization
    return img
```

```python
X = []
y = []

for cls in classes:
    folder = os.path.join("dataset", cls)
    
    for img_name in os.listdir(folder):
        img_path = os.path.join(folder, img_name)
        
        try:
            img = preprocess_image(img_path)
            X.append(img)
            y.append(label_dict[cls])
        except:
            print("Skipped:", img_path)

X = np.array(X)
y = np.array(y)

print("Dataset shape:", X.shape)
print("Labels shape:", y.shape)
```

```python
y_onehot = to_categorical(y, num_classes=4)
print("One-hot shape:", y_onehot.shape)
```

```python
print(y_onehot[:10])
```

```python
datagen = ImageDataGenerator(
    rotation_range=20,
    zoom_range=0.2,
    width_shift_range=0.10,
    height_shift_range=0.10,
    horizontal_flip=True
)
```

```python
sample = X[0]  # first image
sample = np.expand_dims(sample, axis=0)

aug_iter = datagen.flow(sample, batch_size=1)

plt.figure(figsize=(10,5))
for i in range(5):
    batch = next(aug_iter)
    img = batch[0]
    plt.subplot(1,5,i+1)
    plt.imshow(img)
    plt.axis("off")
```

```python
augmented_images = []
augmented_labels = []

for i in range(len(X)):
    img = np.expand_dims(X[i], axis=0)
    label = y_onehot[i]
    
    # Generate 2 augmented images for each
    for _ in range(2):
        # Use next() function instead of .next() method
        aug = next(datagen.flow(img, batch_size=1))[0]
        augmented_images.append(aug)
        augmented_labels.append(label)

augmented_images = np.array(augmented_images)
augmented_labels = np.array(augmented_labels)

print("Augmented:", augmented_images.shape)
```

```python
X_final = np.concatenate([X, augmented_images])
y_final = np.concatenate([y_onehot, augmented_labels])

print("Final dataset:", X_final.shape)
print("Final labels:", y_final.shape)
```
